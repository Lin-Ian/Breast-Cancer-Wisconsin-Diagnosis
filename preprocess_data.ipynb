```python
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('data/cleaned_data.csv')
x = data.drop(['diagnosis'], axis=1)
y = data['diagnosis']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x_train_columns = x.columns
x_train = x_train
min_max_scaler = MinMaxScaler()
x_train_scaled = min_max_scaler.fit_transform(x_train)
x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train_columns)
x_train_scaled
```

```python
from pickle import dump
```

```python
dump(min_max_scaler, open('saved_objects/scaler.pkl', 'wb'))
```

```python
y_train = y_train.map({'M': 1, 'B': 0})
y_train
```

```python
processed_data = pd.concat([pd.DataFrame(y_train.reset_index(drop=True)), x_train_scaled], axis=1)
```

```python
plt.figure(figsize=(24,12))

sns.heatmap(processed_data.corr(),cmap='BrBG',fmt='.2f',
			linewidths=2,annot=True)
plt.savefig('visuals/correlation_plot/correlations_after_processing.png', bbox_inches='tight')
```
